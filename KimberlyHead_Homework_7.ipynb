```python
# Create a function that runs a file and checks that each student passed every test and give
# a count of how many passed
def passed(filename):
    with open(filename, 'r') as f:
        count = 0
        for line in f:      # read through each line
            line = line.strip()     # remove all of the extra characters or whitespace at start and end
            name, grade1, grade2, grade3 = line.split()    # give an identification for each part
            if int(grade1) >= 70 and int(grade2) >= 70 and int(grade3) >= 70:    # Check that each grade passes using 'and'
                count += 1
    print(f'{count} Student\'s Passed')

passed('grades.txt')
```

```python
# Create a function that determines what users were logged in for more than
# 60 minutes
def logins_extendedtime(filename):
    with open(filename, 'r') as f:
        for line in f:
            line=line.strip()
            name, login, logout = line.split(',')
            login_hours, login_minutes = login.split(':')    # Break down all parts
            logout_hours, logout_minutes = logout.split(':')

            login_time = (int(login_hours) * 60) + int(login_minutes)   # turn the hours into minutes
            logout_time = (int(logout_hours) * 60) + int(logout_minutes)  # for easy calculation
            if logout_time - login_time >= 60:
                print(f'{name} Logged in for {logout_time - login_time} minutes')
logins_extendedtime('logfile.txt')
```

```python
# Create a function to take an input file and format it with capitalizing first and last name, and adding area code (301)
# to an output file
def personal_info_formatting(file_in, file_out):
    with open(file_in, 'r') as f_in, open(file_out, 'w') as f_out:
        formatted = [
            (
                f"{' '.join([part.capitalize() for part in line.strip().split('\t')[0].split()])}\t"
                f"{line.strip().split('\t')[1]}\t"
                f"301-{line.strip().split('\t')[2]}"
            )
            for line in f_in
        ]
        for entry in formatted:
            print(entry, file=f_out)
    with open(file_out, 'r') as f_out:
        print(f_out.read())

personal_info_formatting('students.txt', 'students2.txt')
```
