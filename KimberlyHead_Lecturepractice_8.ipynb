```python
# Create a function that add the digits of a number
def sum_digits(num):
    return sum(int(number) for number in str(num))

sum_digits(279)

```

```python
# Using the math.sin function that only takes radians,
# create a function that allows the sine function to work with degrees
import math

def sin_degrees(degrees):
    radians = math.pi * degrees / 180
    return math.sin(radians)

sin_degrees(90)
```

```python
# Create a function the solves an equation and finds all poosible x and y values
def solve_system(a, b, c, d, e, f):
  denominator = (a * d -b * c)
  if denominator == 0:
    raise ValueError("No unique solution")
  x = (d * e - b * f) / denominator
  y = (a * f - c * e) / denominator
  return [x, y]

solve_system(10, 12, 30, 4, 15, 60)

```
